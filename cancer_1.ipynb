CANCER DETECTION MODEL

```python
# import necessary libraries
import numpy as np
import pandas as pd
```

1. Get Data

```python
data = pd.read_csv('survey_lung_cancer.csv')
data.head()
```

2. Preprocessing

```python
# Checking for null values
data.isnull().values.any()
```

```python
data.head()
```

```python
data.shape
```

    a. Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data["LUNG_CANCER"] = encoder.fit_transform(data["LUNG_CANCER"])
data.head()
```

```python
data["SMOKING"] = encoder.fit_transform(data["SMOKING"])
data.head()
```

```python
data["YELLOW_FINGERS"] = encoder.fit_transform(data["YELLOW_FINGERS"])
data.head()
```

```python
data["ANXIETY"] = encoder.fit_transform(data["ANXIETY"])
data.head()
```

```python
data["PEER_PRESSURE"] = encoder.fit_transform(data["PEER_PRESSURE"])
data.head()
```

```python
data["CHRONIC DISEASE"] = encoder.fit_transform(data["CHRONIC DISEASE"])
data.head()
```

```python
data["FATIGUE "] = encoder.fit_transform(data["FATIGUE "])
data.head()
```

```python
data["ALLERGY "] = encoder.fit_transform(data["ALLERGY "])
data.head()
```

```python
data["WHEEZING"] = encoder.fit_transform(data["WHEEZING"])
data.head()
```

```python
data["ALCOHOL CONSUMING"] = encoder.fit_transform(data["ALCOHOL CONSUMING"])
data.head()
```

```python
data["COUGHING"] = encoder.fit_transform(data["COUGHING"])
data.head()
```

```python
data["SHORTNESS OF BREATH"]=encoder.fit_transform(data["SHORTNESS OF BREATH"])
data.head()
```

```python
data["SWALLOWING DIFFICULTY"] = encoder.fit_transform(data["SWALLOWING DIFFICULTY"])
data.head()
```

```python
data["CHEST PAIN"] = encoder.fit_transform(data["CHEST PAIN"])
data.head()
```

```python
data["GENDER"]=encoder.fit_transform(data["GENDER"])
data.head()
```

2. Picking A Model - Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='entropy', max_depth=3)
```

```python
# Splitting input and output
x = data
x = x.drop("LUNG_CANCER", axis=1)
y = data["LUNG_CANCER"]
```

3. Fitting The Model To The Training Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
model.fit(x_train, y_train)
```

4. Evaluate The Model

```python
# Prediction
y_pred = model.predict(x_test)

# Accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_pred, y_test)
accuracy
```

```python
# Classification Report = Precision, Recall, F1 Score
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, target_names=["Cancer", "Not Cancer"]))
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

5. Saving The Model

```python
from joblib import dump
dump(model, 'Lung_Cancer_Detection.joblib')
```
